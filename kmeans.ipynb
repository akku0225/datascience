```
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.datasets import make_blobs


k=3

colors = ["green","red","blue"]
clusters = {}
for i in range(k):


    center=12*(2*np.random.random((2,))-1)
    points = []


    cluster = {
         'center' : center,
         'points' : [],
         'color'  : colors[i]

      }
    clusters[i] = cluster


for i in range(k):
     plt.scatter(clusters[i]['center'][0],clusters[i]['center'][1])
     plt.show()
```

```
def  dist(p1,p2):
     return np.sum((p1-p2)**2)**.5
p1 = np.array([3,6])
p2 = np.array([7,9])
#customers step1
def assigncustomer():
 m= X.shape[0]
 for ix in range(m):
      cdist = []
      for kx in range(k):
          d =dist(X[ix],clusters[kx]['center'])
          cdist.append(d)
      ccluster = np.argmin(cdist)
      clusters[ccluster]['points'].append(X[ix])
#step3
def updatecenters():
    for i in range(k):
         points = np.array(clusters[i]['points'])
         if(points.shape[0]>0):
              clusters[i]['center'] = np.mean(points,axis=0)
              clusters[i]['points'] = []
         
      
```

```
#step 2 plotting
 
def visual():
 for i in range(k):
   plt.scatter(clusters[i]['center'][0],clusters[i]['center'][1],color=clusters[i]['color'],marker='+')
   if len(clusters[i]['points'])>0:
        points = np.array(clusters[i]['points'])
        plt.scatter(points[:,0],points[:,1],color=clusters[i]['color'])


 plt.show()
```

```
assigncustomer()
visual()
updatecenters()
visual()
```

```
from sklearn.cluster import KMeans
```

```
k=4

model = KMeans(n_clusters=k)
model.fit(all_pixal)
```

```
centers = model.cluster_centers_
centers = np.array(centers,dtype='uint8')
centers
```
